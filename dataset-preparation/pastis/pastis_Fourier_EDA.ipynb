# Red

```python
red_converted = convert_to_fourier_basis(data_dir, "red", threshold =0.02, max_depth = 9, presplit= 2, image_func=normalizer)
red_converted["number_of_coefficents"] = red_converted["data"].apply(lambda x:len(x))/(2 *n)
red_converted
```

```python
red_converted = convert_to_fourier_basis(data_dir, "red", threshold =0.03, max_depth = 9, image_func=normalizer)
red_converted["number_of_coefficents"] = red_converted["data"].apply(lambda x:len(x))/(2 *n)
red_converted
```

```python
filtered_red = red_converted[red_converted["magnitude_endpoints"].apply(lambda x: x[0] > 0.0078125)]
filtered_red
```

```python

filtered_red["magnitude_length"] = filtered_red["magnitude_endpoints"].apply(lambda x: x[1] - x[0])
x = np.asfarray(filtered_red["band"].to_numpy())
y = np.asfarray(filtered_red["magnitude_endpoints"].apply(lambda x: x[1]).to_numpy())
logy = np.log(y)
plt.plot(x, logy)
print(np.poly1d(np.polyfit(x, logy, 1)))
plt.plot(x, np.poly1d(np.polyfit(x, logy, 1))(x))
np.exp(np.polyfit(x, logy, 1)[0])
```
