```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randint(50, 89, 80)
bins = np.arange(50, data.max()+6, 5)
fig, ax = plt.subplots(figsize=(12,5))

ax.hist(data, bins=bins, fc='skyblue', ec='navy')
false_zero = bins[0] - 10
ax.set_xticks(np.append(false_zero, bins))
ax.set_xticklabels(np.append(0, bins))

for spine in ['top', 'right', 'bottom']:
    ax.spines[spine].set_visible(False)
ax.add_line(plt.Line2D(xdata=[false_zero, false_zero + 2, false_zero + 4, false_zero + 4, false_zero + 6, bins[-1] + 1],
                       ydata=[0, 0, 0.1, -0.1, 0, 0],
                       color=ax.spines['bottom'].get_edgecolor(), lw=ax.spines['bottom'].get_linewidth(),
                       clip_on=False, transform=ax.get_xaxis_transform()))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate a parabola
x = np.linspace(-10, 10, 500)  # X values
y = -0.1 * (x - 7)**2 + 8      # Parabola shifted right and upward

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 5))

# Plot the parabola
ax.plot(x, y, color='skyblue', lw=3)

# Create custom x-axis with a kink on the right
false_zero = x.min()  # Faking the x-axis origin
kink_x = x.max() - 3  # Position of the kink

ax.add_line(plt.Line2D(
    xdata=[false_zero, kink_x, kink_x + 0.2, kink_x + 0.4, kink_x + 0.6, x.max()],
    ydata=[0, 0, 0.3, -0.3, 0, 0],
    color='black', lw=1.5, clip_on=False, transform=ax.get_xaxis_transform()
))

# Customize the plot appearance
for spine in ['top', 'right', 'bottom']:
    ax.spines[spine].set_visible(False)

# Remove default ticks and add a label to the x-axis
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim(x.min(), x.max())
ax.set_ylim(y.min() - 2, y.max() + 2)
ax.set_xlabel("X-axis", fontsize=12)
ax.set_ylabel("Y-axis", fontsize=12)

# Add arrows to the axes
# X-axis arrow
ax.annotate('', xy=(x.max() + 0.5, 0), xytext=(x.max(), 0),
            arrowprops=dict(facecolor='black', arrowstyle='->'))

# Y-axis arrow
ax.annotate('', xy=(false_zero, y.max() + 2), xytext=(false_zero, y.max()),
            arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.tight_layout()
plt.show()
```
